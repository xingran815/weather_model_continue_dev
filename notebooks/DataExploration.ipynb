```python
# import pandas for using DataFrames and numpy for calculations
import pandas as pd
import numpy as np

# import seaborn and matplotlib for visualisation
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read raw data

#this is Franziskas download path
df = pd.read_csv(r"C:\Users\stefa\Documents\GitHub\oct25_bmlops_int_weather\data\raw\weatherAUS.csv", sep=',', encoding='latin1', decimal='.')

df.head()
```

```python
# display the number of rows and columns of the data
print('The dataset has ', df.shape[1], 'different columns. It contains of ', df.shape[0], 'rows/data entrys.')
print('')

# getting the names and types of data columns
print('Information about the type of columns: ')
display(df.info())

# finding out how many values are missing in each column in %
print('Percentage of missing values in each column (in %): ')
display(df.isna().sum()/len(df)*100)
```

```python
df.describe()
```

```python
# get unique values

for column in df.columns:
    if df[column].dtype=='object':
        display(df[column].value_counts())
        display(df[column].unique())
        display(df[column].unique().shape)
```

```python
# show correlations between numerical columns by Pearson coefficient
fig, ax = plt.subplots(figsize=(12,12))
sns.heatmap(df.corr(numeric_only=True),annot=True ,ax= ax)
```

```python
# show rain
sns.countplot(data=df, x='RainToday')
plt.title('Rain Today')
plt.xticks(rotation=45)
plt.xlabel('');
```

```python
# show rain
sns.countplot(data=df, x='RainTomorrow')
plt.title('Rain Tomorrow')
plt.xticks(rotation=45)
plt.xlabel('');
```

```python
# compare statistical values 
group_hybrids=df.groupby(['RainToday']).agg({'RainTomorrow': ['value_counts']})
display(group_hybrids)
group_hybrids.to_markdown(index=False)
```

```python
# compare statistical values 
group_hybrids=df.groupby(['Location']).agg({'RainToday': ['value_counts']})
display(group_hybrids)
```
